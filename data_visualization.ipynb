# Number of people/frame from pose data

import os
import json
import collections
inputdir = '/mnt/disks/disk1/raw/densepose/txt'
max_ppl = 0
counts = {}
for d in os.listdir(inputdir):
    this_max_ppl = 0
    counts[d] = collections.Counter()
    for fin in os.listdir(os.path.join(inputdir, d)):
        fin = os.path.join(inputdir, d, fin)
        f = json.load(open(fin))
        num_ppl = len(f)
        counts[d][num_ppl] += 1
        if num_ppl > this_max_ppl:
            this_max_ppl = len(f)
    print("finished {}, this_max_ppl {}".format(d, this_max_ppl))

```python
import matplotlib.pyplot as plt
%matplotlib inline
for d in counts: 
    plt.bar(list(counts[d].keys()), counts[d].values(), color='g')
    plt.title("{} number of people".format(d))
    plt.xlabel("number people")
    plt.ylabel("instances")
    plt.show()
```
